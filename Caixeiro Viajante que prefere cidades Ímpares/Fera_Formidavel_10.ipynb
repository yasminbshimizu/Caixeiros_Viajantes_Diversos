Agora sim podemos come√ßar a trabalhar. Esse problema se assemelha muito ao problema do caixeiro viajante comum, j√° resolvido em aula. Por isso, podemos nos basear nele e apenas implementar algumas modifica√ß√µes. Dessa forma, um indiv√≠duo ser√° uma lista contendo os n√∫meros referentes √†s cidades que o caixeiro ir√° visitar na ordem em que ele o far√°. Assim, podemos reutilizar as fun√ß√µes: <cpde>funcao_objetivo_caixeiro</code>, <code>funcao_objetivo_pop_caixeiro</code> e <code>selecao_torneio_min</code>. </div>

<div style=" padding: 10px; text-align: justify">
Vamos, ent√£o, ao que precisamos mudar. J√° que precisamos atribuir um n√∫mero inteiro a cada cidade, devemos fazer uma pequena modifica√ß√£o na fun√ß√£o <code>cria_cidades</code>. Para isso, criaremos a fun√ß√£o <code>cria_cidades_impar</code>. Essa fun√ß√£o receber√° $n$, um n√∫mero inteiro representando o n√∫mero de cidades a serem geradas, <code>xy_minimo=0</code>, reepresentando o menor valor poss√≠vel para as coordenadas $(x, y)$, e <code>xy_maximo=300</code>, reepresentando o maior valor poss√≠vel para as coordenadas $(x, y)$. Podemos verificar o funcionamento dessa fun√ß√£o na c√©lula abaixo.</div>

```python
import random

def cria_cidades_impar(n, xy_minimo=0, xy_maximo=300):
    """Cria um dicion√°rio aleat√≥rio de cidades com suas posi√ß√µes (x,y), com a chave do docion√°rio sendo um n√∫mero inteiro.

    Args:
      n: N√∫mero de cidades que ser√£o visitadas pelo caixeiro.
      xy_minimo: Valor m√≠nimo poss√≠vel das coordenadas x e y.
      xy_maximo: Valor m√°ximo poss√≠vel das coordenadas x e y.

    """
    cidades = {}

    for i in range(n):
        cidades[i] = (
            random.randint(xy_minimo, xy_maximo),
            random.randint(xy_minimo, xy_maximo),
        )

    return cidades

print("Cidades geradas:")
cria_cidades_impar(10)
```

<div style=" padding: 10px; text-align: justify">
O enunciado diz, tamb√©m, ser necess√°rio que o caixeiro visite primeiro cidades √≠mpares, ent√£o um indiv√≠duo s√≥ √© v√°lido se come√ßar com 0, listar todas as cidades √≠mpares e, somente ent√£o, listar as cidades pares. A fun√ß√£o <code>cria_candidato_caixeiro</code> n√£o satisfaz essa restri√ß√£o, ent√£o precisamos criar uma nova, que chamaremos <code>cria_candidato_caixeiro_impar</code>. Essa fun√ß√£o recebe o dicion√°rio de cidades (criado na fun√ß√£o anterior), retira a cidades 0 (pois ela ser√° a primeira em qualquer candidato) e embaralha as cidades, formando um caminho aleat√≥rio. Em seguida, na ordem em que aparecem, separa as cidades √≠mpares em uma lista e as cidades pares em outra. Por fim, cria o caminho final, come√ßando com 0, adicionando as cidades √≠mpares e, por √∫ltimo, as cidades pares. Esse processo garante que o indiv√≠duo ser√° v√°lido.</div>

<div style=" padding: 10px; text-align: justify">
Como precisamos de mais de um indiv√≠duo para a popula√ß√£o, temos que fazer tamb√©m a fun√ß√£o <code>populacao_caixeiro_impar</code>, que simplesmente chama <code>cria_candidato_caixeiro_impar</code> um n√∫mero de vezes igual ao tamanho da popula√ß√£o estabelecido.</div>

```python
def cria_candidato_caixeiro_impar(cidades):
    """Sorteia um caminho poss√≠vel no problema do caixeiro viajante que prefere cidades √≠mpares.

    Args:
      cidades:
        Dicion√°rio contendo o nome das cidades como chaves e a coordenada no
        plano cartesiano das cidades como valores.

    """
    
    caminho_0 = [0]

    cidades_menos0 = list(cidades.keys())
    cidades_menos0.remove(0)
    
    caminho_menos0 = random.sample(cidades_menos0, k=len(cidades_menos0))
    
    caminho_impar = [i for i in caminho_menos0 if i%2 != 0]
    caminho_par = [i for i in caminho_menos0 if i%2 == 0]
   
    caminho = caminho_0 + caminho_impar + caminho_par
    
    return caminho

def populacao_caixeiro_impar(tamanho_populacao, cidades):
    """Cria uma popula√ß√£o no problema do caixeiro viajante que prefere cidades √≠mpares.

    Args:
      tamanho_populacao: tamanho da popula√ß√£o.
      cidades:
        Dicion√°rio contendo o nome das cidades como chaves e a coordenada no
        plano cartesiano das cidades como valores.

    """
    populacao = []

    for _ in range(tamanho_populacao):
        populacao.append(cria_candidato_caixeiro_impar(cidades))

    return populacao

print("Popula√ß√£o:\n\n")

populacao = populacao_caixeiro_impar(10, cria_cidades_impar(10))
for n, individuo in enumerate(populacao):
    print(f"Indiv√≠duo {n+1}: {individuo}\n")
```

<div style=" padding: 10px; text-align: justify">
Outra fun√ß√£o que precisaremos mudar, para garantir que n√£o iremos gerar indiv√≠duos inv√°lidos, √© a fun√ß√£o de cruzamento. O que faremos, nesse caso, √© apenas criar uma fun√ß√£o intermedi√°ria que chamar√° a fun√ß√£o de cruzamento ordenado, j√° feita em aula. Esse operador, basicamente, faz o cruzamento garantindo que n√£o haver√° elementos repetidos em um indiv√≠duo. Precisamos certificar, por√©m, que o primeiro elemento continue sendo 0 e que pares e √≠mpares n√£o se misturem. </div>

<div style=" padding: 10px; text-align: justify">
Para isso, primeiro criamos a fun√ß√£o <code>calc_min_idx_par</code>, que, ao receber o dicion√°rio de cidades, retorna um inteiro representando o primeiro √≠ndice da lista que pode ser par (ou seja, elementos de √≠ndices menores devem ser 0 ou √≠mpares, enquanto elementos de √≠ndices maiores ou iguais devem ser pares). Ent√£o, criamos a fun√ß√£o <code>cruzamento_ordenado_intervalos</code>, a qual realiza o cruzamento ordenado (usando a fun√ß√£o <code>cruzamento_ordenado</code>) em intervalos, ou seja, primeiro cruza os elementos pares e depois os elementos √≠mpares (que descobre usando <code>calc_min_idx_par</code>). Depois de cruzar os intervalos separados, basta junt√°-los para retornar os filhos. </div>

```python
def calc_min_idx_par(cidades):
    """Calcula o √≠ndice do dicion√°rio a partir do qual todas as cidades s√£o pares.

    Args:
      cidades:
        Dicion√°rio contendo o nome das cidades como chaves e a coordenada no
        plano cartesiano das cidades como valores.

    """
    min_idx_par = len(list(cidades.keys()))//2 + 1
    return min_idx_par


def cruzamento_ordenado(pai, mae, chance_de_cruzamento):
    """Cruzamento ordenado entre dois individuos

    Args:
      pai: lista representando um individuo
      mae: lista representando um individuo
      chance_de_cruzamento: float entre 0 e 1 representando a chance de cruzamento

    """
    if random.random() < chance_de_cruzamento:
        tamanho_individuo = len(mae)

        # pontos de corte
        corte1 = random.randint(0, tamanho_individuo - 2)
        corte2 = random.randint(corte1 + 1, tamanho_individuo)

        # filho1
        filho1 = [None] * tamanho_individuo
        filho1[corte1:corte2] = mae[corte1:corte2]
        pai_ = pai[corte2:] + pai[:corte2]
        posicao = corte2 % tamanho_individuo
        for valor in pai_:
            if valor not in filho1:
                filho1[posicao] = valor
                posicao += 1
                posicao %= tamanho_individuo

        # filho2
        filho2 = [None] * tamanho_individuo
        filho2[corte1:corte2] = pai[corte1:corte2]
        mae_ = mae[corte2:] + mae[:corte2]
        posicao = corte2 % tamanho_individuo
        for valor in mae_:
            if valor not in filho2:
                filho2[posicao] = valor
                posicao += 1
                posicao %= tamanho_individuo

        return filho1, filho2
    else:
        return pai, mae


def cruzamento_ordenado_intervalos(pai, mae, chance_de_cruzamento, cidades):
    """Cruzamento ordenado entre dois individuos dividido em dois intervalos (√≠mpares e pares) para o problema do caixeiro viajante que prefere cidades √≠mpares.
 
    Args:
      pai: lista representando um individuo
      mae: lista representando um individuo
      chance_de_cruzamento: float entre 0 e 1 representando a chance de cruzamento
      cidades:
        Dicion√°rio contendo o nome das cidades como chaves e a coordenada no
        plano cartesiano das cidades como valores.
 
    """
    if random.random() < chance_de_cruzamento:
        min_idx_par = calc_min_idx_par(cidades)
 
        mae_impar = mae[1:min_idx_par]
        pai_impar = pai[1:min_idx_par]
 
        mae_par = mae[min_idx_par:]
        pai_par = pai[min_idx_par:]
 
        filho1_impar, filho2_impar = cruzamento_ordenado(pai_impar, mae_impar, 1)
        filho1_par, filho2_par = cruzamento_ordenado(pai_par, mae_par, 1)
 
        filho1 = [0] + filho1_impar + filho1_par
        filho2 = [0] + filho2_impar + filho2_par
        
        return filho1, filho2
    else:
        return pai, mae
    
cidades = cria_cidades_impar(10)
    
pai = cria_candidato_caixeiro_impar(cidades)
mae = cria_candidato_caixeiro_impar(cidades)

filho1, filho2 = cruzamento_ordenado_intervalos(pai, mae, 1, cidades)

print(f"Pai: {pai}")
print(f"Mae: {mae}")
print()
print(f"Filho 1: {filho1}")
print(f"Filho 2: {filho2}")
```

<div style=" padding: 10px; text-align: justify">
Por fim, criamos a fun√ß√£o <code>mutacao_troca_cx_impar</code> para fazer a muta√ß√£o dos indiv√≠duos. Ela √© baseada no operador de muta√ß√£o de troca, que troca a posi√ß√£o de dois elementos de um indiv√≠duo. A diferen√ßa desse √© que ele confere se o elemento escolhido √© par ou √≠mpara e garante que o outro elemento seja da mesma paridade (e nunca escolhe 0). </div>

```python
def mutacao_troca_cx_impar(populacao, chance_de_mutacao, cidades):
    """Aplica mutacao de troca em um indiv√≠duo no problema do caixeiro viajante que prefere cidades √≠mpares.

    Args:
      populacao: lista contendo os indiv√≠duos do problema
      chance_de_mutacao: float entre 0 e 1 representando a chance de muta√ß√£o
      cidades:
        Dicion√°rio contendo o nome das cidades como chaves e a coordenada no
        plano cartesiano das cidades como valores.
    """
    
    min_idx_par = calc_min_idx_par(cidades)
    
    for individuo in populacao:
        if random.random() < chance_de_mutacao:
            gene1 = random.randint(1, len(individuo) - 1)
            if gene1 < min_idx_par:
                gene2 = random.randint(1, min_idx_par - 1)
            else:
                gene2 = random.randint(min_idx_par, len(individuo) - 1)
                
            while gene1 == gene2:
                gene1 = random.randint(1, len(individuo) - 1)
                gene2 = random.randint(1, len(individuo) - 1)
                
                if gene1 < min_idx_par:
                    gene2 = random.randint(1, min_idx_par - 1)
                else:
                    gene2 = random.randint(min_idx_par, len(individuo) - 1)

            individuo[gene1], individuo[gene2] = (
                individuo[gene2],
                individuo[gene1],
            )
```

### Evoluindo o algoritmo gen√©tico

Com as fun√ß√µes prontas, podemos evoluir o algoritmo gen√©tico e, com efeito, resolver o problema. Vamos importar as fun√ß√µes necess√°rias.

```python
from functools import partial
from itertools import permutations

from funcoes_fera_10 import cria_cidades_impar as cria_cidades
from funcoes_fera_10 import populacao_caixeiro_impar as cria_populacao
from funcoes_fera_10 import funcao_objetivo_pop_caixeiro
from funcoes_fera_10 import funcao_objetivo_caixeiro
from funcoes_fera_10 import selecao_torneio_min as funcao_selecao
from funcoes_fera_10 import cruzamento_ordenado_intervalos
from funcoes_fera_10 import mutacao_troca_cx_impar
```

Agora, definimos as constantes do problema.

```python
NUM_CIDADES = 15
CIDADES = cria_cidades(NUM_CIDADES)

TAMANHO_POPULACAO = 100
NUM_GERACOES = 500
CHANCE_DE_CRUZAMENTO = 0.5
CHANCE_DE_MUTACAO = 0.05
TAMANHO_TORNEIO = 3
```

Criamos a poupula√ß√£o inicial.

```python
populacao = cria_populacao(TAMANHO_POPULACAO, CIDADES)
# pprint(populacao)
# funcao_objetivo(populacao, CIDADES)
```

E definimos algumas fun√ß√µes parciais para adicionar par√¢metros necess√°rios √†s fun√ß√µes.

```python
funcao_objetivo = partial(funcao_objetivo_pop_caixeiro, cidades=CIDADES)
funcao_cruzamento = partial(cruzamento_ordenado_intervalos, cidades=CIDADES)
funcao_mutacao = partial(mutacao_troca_cx_impar, cidades=CIDADES)
```

Por fim, rodamos o algoritmo.

```python
hall_da_fama = []

for n in range(NUM_GERACOES):
    
    # Sele√ß√£o
    fitness = funcao_objetivo(populacao)        
    selecionados = funcao_selecao(populacao, fitness, TAMANHO_TORNEIO)
    
    # Cruzamento
    proxima_geracao = []
    for pai, mae in zip(selecionados[::2], selecionados[1::2]):
        individuo1, individuo2 = funcao_cruzamento(pai, mae, CHANCE_DE_CRUZAMENTO)
        proxima_geracao.append(individuo1)
        proxima_geracao.append(individuo2)
    
    # Muta√ß√£o
    funcao_mutacao(proxima_geracao, CHANCE_DE_MUTACAO)
    
    # Atualiza√ß√£o do hall da fama
    fitness = funcao_objetivo(proxima_geracao)
        
    menor_fitness = min(fitness)
    indice = fitness.index(menor_fitness)
    hall_da_fama.append(proxima_geracao[indice])    
    
    # Encerramento
    populacao = proxima_geracao
```

Podemos visualizar o melhor resultado encontrado.

```python
fitness = funcao_objetivo(hall_da_fama)
menor_fitness = min(fitness)
indice = fitness.index(menor_fitness)
melhor_individuo_observado = hall_da_fama[indice]
distancia = funcao_objetivo_caixeiro(melhor_individuo_observado, CIDADES)

print()
print("Melhor individuo obtido por algoritmos gen√©ticos:")
print(melhor_individuo_observado, "com dist√¢ncia:", distancia)
print()

plota_trajeto(CIDADES, melhor_individuo_observado)
```
